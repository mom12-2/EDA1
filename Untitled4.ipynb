```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import xgboost as xgb
import shap
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import log_loss, make_scorer, classification_report, average_precision_score

# ==========================================
# 1. GENERATE SYNTHETIC DATASET (FALLBACK)
# ==========================================
print("--- STEP 1: Setting up Dataset ---")

# We simulate a credit card fraud dataset: 50,000 transactions, 9 features, 0.5% fraud rate
X_raw, y_raw = make_classification(
    n_samples=50000,
    n_features=9,
    n_informative=7,
    n_redundant=2,
    weights=[0.995, 0.005],  # 0.5% fraud mimicking severe imbalance
    random_state=42
)

# Convert to DataFrame to mimic your selected columns file structure
feature_names = [f'V{i}' for i in range(1, 9)] + ['Amount']
df = pd.DataFrame(X_raw, columns=feature_names)
df['Class'] = y_raw

print(f"Dataset generated successfully with shape: {df.shape}")
print(f"Class distribution:\n{df['Class'].value_counts(normalize=True) * 100}\n")

# Separate Features and Target
y = df['Class'].astype(int)
X = df.drop(columns=['Class'])

# Scale the 'Amount' feature
scaler = StandardScaler()
X['Amount'] = scaler.fit_transform(X[['Amount']])

# Split into stratified Train and Validation sets
X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(f"Training shape: {X_train.shape}, Validation shape: {X_val.shape}")


# ==========================================
# 2. SKLEARN GRADIENT BOOSTING & OVERFITTING
# ==========================================
print("\n--- STEP 2: Running Sklearn Gradient Boosting & GridSearchCV ---")

param_grid = {
    'n_estimators': [30, 60],
    'learning_rate': [0.05, 0.2]
}

log_loss_scorer = make_scorer(log_loss, greater_is_better=False, response_method='predict_proba')

gb_grid = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid,
    scoring=log_loss_scorer,
    cv=3,
    n_jobs=-1
)
gb_grid.fit(X_train, y_train)
print(f"Best Hyperparameters Found: {gb_grid.best_params_}")

# Evaluate overfitting trends with a learning rate across 100 estimators
best_lr = gb_grid.best_params_['learning_rate']
max_estimators = 100

gb_model = GradientBoostingClassifier(n_estimators=max_estimators, learning_rate=best_lr, random_state=42)
gb_model.fit(X_train, y_train)

train_errors = []
val_errors = []

for y_pred_train, y_pred_val in zip(gb_model.staged_predict_proba(X_train), gb_model.staged_predict_proba(X_val)):
    train_errors.append(log_loss(y_train, y_pred_train))
    val_errors.append(log_loss(y_val, y_pred_val))

# Plot the training vs validation error curves
plt.figure(figsize=(10, 4))
plt.plot(range(1, max_estimators + 1), train_errors, label='Training Error', color='blue')
plt.plot(range(1, max_estimators + 1), val_errors, label='Validation Error', color='red')
plt.xlabel('Number of Estimators')
plt.ylabel('Log Loss')
plt.title(f'Gradient Boosting Error Curves (Learning Rate = {best_lr})')
plt.legend()
plt.grid(True)
plt.show()


# ==========================================
# 3. XGBOOST WITH EARLY STOPPING & BALANCING (FIXED API)
# ==========================================
print("\n--- STEP 3: Training XGBoost Classifier ---")

# Calculate class weights dynamically to fix data imbalance
num_neg = np.sum(y_train == 0)
num_pos = np.sum(y_train == 1)
scale_weight_value = num_neg / max(1, num_pos)
print(f"Calculated scale_pos_weight: {scale_weight_value:.2f}")

# FIX: Move early_stopping_rounds here into the initialization
xgb_model = xgb.XGBClassifier(
    n_estimators=500,
    learning_rate=0.05,
    scale_pos_weight=scale_weight_value,
    eval_metric='logloss',
    early_stopping_rounds=15,    # <--- Added here
    random_state=42,
    n_jobs=-1
)

# FIX: Remove early_stopping_rounds from the fit parameters
xgb_model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=25                  # <--- Kept clean here
)

# Performance metrics output
y_pred = xgb_model.predict(X_val)
print("\nXGBoost Performance Metrics Summary:")
print(classification_report(y_val, y_pred))
print(f"Area Under Precision-Recall Curve (PR-AUC): {average_precision_score(y_val, xgb_model.predict_proba(X_val)[:, 1]):.4f}")


# ==========================================
# 4. SHAP EXPLANATIONS FOR INTERPRETABILITY
# ==========================================
print("\n--- STEP 4: Calculating SHAP Feature Importance ---")

# Compute SHAP values using the TreeExplainer
explainer = shap.TreeExplainer(xgb_model)
X_val_sample = X_val.sample(n=300, random_state=42)
shap_values = explainer(X_val_sample)

# Generate summary plot
plt.figure(figsize=(10, 5))
shap.summary_plot(shap_values, X_val_sample, show=False)
plt.title("SHAP Feature Importance Summary Plot", fontsize=14)
plt.tight_layout()
plt.show()
```

### Part 1: Overfitting and Learning Rates

#### **1. When does Gradient Boosting start to overfit?**

Gradient Boosting starts to overfit at the **global minimum of the Validation Error curve (Log Loss)**.

Before this point, both training and validation errors decrease together (generalization). After this point, a clear divergence occurs: the training error continues to decline toward zero, while the validation error bottoms out and begins to trend upward. This inflection point means the model has stopped learning generalizable patterns and has begun memorizing the noise, outliers, and variance unique to the training data.

#### **2. How does the learning rate affect this point?**

The learning rate ($\eta$) acts as a scaling factor for each tree's contribution. It changes the overfitting point in two ways:

* **High Learning Rate (e.g., 0.20):** Base learners make aggressive corrections. The model converges rapidly, **shifting the overfitting point to the left** (occurring early, at a low number of estimators). The validation loss curve forms a sharp, steep "V" shape.
* **Low Learning Rate (e.g., 0.05):** Base learners make minor, conservative updates. This slows down convergence and **shifts the overfitting point to the right** (delayed until a high number of estimators). This forms a flat, stable "U" validation curve that is much more resilient to overfitting.

### Part 2: Advanced Boosting (XGBoost) & SHAP

#### **1. What is the role of Early Stopping?**

Instead of blindly training a fixed number of trees, **Early Stopping** monitors out-of-sample validation loss at each iteration. If the validation loss fails to improve for a set window (e.g., 15 consecutive rounds), training automatically halts. This dynamically catches the optimal validation minimum before overfitting begins, saving computational resources and ensuring peak model generalization.

#### **2. How do you handle class imbalance using `scale_pos_weight`?**

In highly imbalanced datasets (like 0.5% fraud), standard accuracy is deceptive because a dummy model guessing "never fraud" yields 99.5% accuracy while catching nothing. Setting `scale_pos_weight` equal to the ratio of negative-to-positive instances:

$$\text{scale\_pos\_weight} = \frac{\text{Total Legitimate Cases}}{\text{Total Fraud Cases}}$$

multiplies the gradient loss penalty whenever the model misclassifies a positive (fraud) case. This shifts the internal decision boundary, forcing the algorithm to prioritize minority class classification and directly optimizing the **Area Under the Precision-Recall Curve (PR-AUC)**.

#### **3. How do SHAP values interpret the model's predictions?**

Unlike traditional feature importance metrics that only count how often a variable is used to split a tree, **SHAP (SHapley Additive exPlanations)** uses cooperative game theory to measure the exact additive marginal contribution of each feature to a specific prediction.

On a **SHAP Summary Plot**:

* **Y-Axis Position:** Features are ranked from top to bottom by total absolute impact; the topmost feature is the strongest model driver.
* **X-Axis Displacement:** Points to the right of the center line indicate features pushing the probability toward a fraud prediction; points to the left pull it toward a legitimate classification.
* **Color Temperature:** Red indicates a high feature value, and blue indicates a low feature value. For example, a cluster of red dots far to the right proves that **high values of that feature directly increase the likelihood of fraud**.
