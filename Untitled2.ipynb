The ROC Curve and AUC
Dataset: from sklearn.datasets import load_breast_cancer
Task: Train a classifier that outputs probabilities (like Logistic Regression or Random Forest) on a binary classification dataset.
Action: Extract the predicted probabilities (predict_proba) rather than the absolute class predictions. Use these probabilities to plot a Receiver Operating Characteristic (ROC) curve. Calculate and display the Area Under the Curve (AUC) score on the plot. Explain what the curve represents regarding True Positive vs. False Positive trade-offs.

```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# 1. Load data and split
data = load_breast_cancer()
X, y = data.data, data.target # 0 = Malignant, 1 = Benign

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# 2. Train model and get predictions
model = LogisticRegression(max_iter=10000, random_state=42)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

# 3. Generate Confusion Matrix
cm = confusion_matrix(y_test, y_pred)
print("Confusion Matrix:")
print(cm)

# Extract individual components
# scikit-learn's confusion matrix structure for binary classification:
# [ [TN, FP],
#   [FN, TP] ]
TN, FP, FN, TP = cm.ravel()
print(f"\nExtracted: TN={TN}, FP={FP}, FN={FN}, TP={TP}")
```

```python
# 4. Manual Calculations
manual_accuracy = (TP + TN) / (TP + TN + FP + FN)
manual_precision = TP / (TP + FP)
manual_recall = TP / (TP + FN)
manual_f1 = 2 * (manual_precision * manual_recall) / (manual_precision + manual_recall)

print("--- Manual Calculations ---")
print(f"Accuracy:  {manual_accuracy:.4f}")
print(f"Precision: {manual_precision:.4f}")
print(f"Recall:    {manual_recall:.4f}")
print(f"F1-Score:  {manual_f1:.4f}\n")

# 5. Verification via Scikit-Learn
print("--- Sklearn Classification Report ---")
print(classification_report(y_test, y_pred, target_names=data.target_names))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, roc_auc_score

# 1. Get predicted probabilities for the positive class (class 1: Benign)
y_probs = model.predict_proba(X_test)[:, 1]

# 2. Calculate True Positive Rate (TPR) and False Positive Rate (FPR) for various thresholds
fpr, tpr, thresholds = roc_curve(y_test, y_probs)

# 3. Calculate Area Under the Curve (AUC)
auc_score = roc_auc_score(y_test, y_probs)

# 4. Plotting
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC Curve (AUC = {auc_score:.4f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--', label='Random Guessing (AUC = 0.5)')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)')
plt.ylabel('True Positive Rate (Sensitivity / Recall)')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True, alpha=0.3)
plt.show()
```
